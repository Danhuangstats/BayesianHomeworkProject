## 1. Rinderpest Virus in Rabbits with Missing Data.

```python
import pandas as pd 
import numpy as np 
import pymc3 as pm
import arviz as az
```

```python
from pymc3.distributions import Normal
```

```python
x=np.array([24,32,48,56,0,70,72,75,80,96])
y=np.array([102.8,104.5,106.5,107.0,107.1,105.1,103.9,0,103.2,102.1])
```

## part a


```python

linemodel=pm.Model()


with linemodel:
    # define priors
    a = pm.Normal('slope', 0, 1)
    b = pm.Normal('intercept',0, 100)
    s = pm.Gamma('error', 0.01,0.01)
    
    # predictions
    obs = pm.Normal('obs', a*x + b, s, observed=y)
    
    
    # use MCMC to sample
    trace = pm.sample(draws=1000, return_inferencedata=True,model=linemodel)

```

```python
summary_df = pm.summary(trace)
summary_df

sst=np.sum((y-y.mean())**2)

ppc=pm.sample_posterior_predictive(trace,samples=1000,model=linemodel)

pred=pd.DataFrame(ppc['obs'])

ypred=[]
for i in range(len(x)):
    predMean=round(pred.iloc[:,i].mean(),2)
    ypred.append(predMean)
    
BR2=1-np.sum((ypred-y)**2)/sst

BR2
```

```python
summary_df = pm.summary(trace)
summary_df
```

```python
pm.plot_posterior(trace,hdi_prob=0.95)
```

## part b

```python
quadmodel=pm.Model()


with quadmodel:
    # define priors
     
    b0= pm.Normal('b0', 0, 1)
    b1= pm.Normal('b1', 0, 1)
    alpha = pm.Normal('alpha',y.mean(), 1)
    s = pm.HalfCauchy('error', 1)
    
    # predictions
    obs = pm.Normal('obs', b0*x + b1 * x**2 + alpha, s, observed=y)
    
    
    # use MCMC to sample
    trace = pm.sample(draws=1000, return_inferencedata=True, model=quadmodel)
```

```python
sst=np.sum((y-y.mean())**2)

ppc=pm.sample_posterior_predictive(trace,samples=1000,model=quadmodel)

pred=pd.DataFrame(ppc['obs'])

ypred=[]
for i in range(len(x)):
    predMean=round(pred.iloc[:,i].mean(),2)
    ypred.append(predMean)
    
BR2=max(0,1-np.sum((ypred-y)**2)/sst)

BR2
```

```python
summary_df = pm.summary(trace)
summary_df
```

```python
pm.plot_posterior(trace,hdi_prob=0.95)
```
